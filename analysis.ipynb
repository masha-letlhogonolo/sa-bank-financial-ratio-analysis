```python
#Data cleaning
import pandas as pd
import numpy as np

fsr = pd.read_csv("FSRJ_fund.csv")
sbk = pd.read_csv("SBKJ_fund.csv")
ned = pd.read_csv("NEDJ_fund.csv")
print(fsr.head())
print(fsr.columns)

#COMBINE THE COMPANIES
fsr["Bank"] = "FSR"
sbk["Bank"] = "SBK"
ned["Bank"] = "NED"
companies = pd.concat([fsr,sbk,ned]) #master dataset
print(companies)

#date conversion
companies["Date"] = pd.to_datetime(companies["Date"], dayfirst=True)
companies = companies.sort_values("Date")
```

```python
#Financial Ratios

#totral assets and liabilities
companies["Total_Assets"] = companies["Non-current assets"] + companies["Current assets"]
companies["Total_Liabilities"] = companies["Non-current liabilities"] + companies["Current liabilities"]

#Profitability
#profit margin
companies["Profit_Margin"] = companies["Net profit"] / companies["Revenue"]
#gross margin
companies["Gross_Margin"] = companies["Gross profit"] / companies["Revenue"]
#return on asset
companies["ROA"] = companies["Net profit"] / companies["Total_Assets"]

#Leverage ratio
companies["Debt_Ratio"] = companies["Total_Liabilities"] / companies["Total_Assets"]
#Liquidity Ratio
companies["Current_Ratio"] = companies["Current assets"] / companies["Current liabilities"]

#average profit margin per bank
print(companies.groupby("Bank")["Profit_Margin"].mean())
#average return on assets
print(companies.groupby("Bank")["ROA"].mean())
# average debt ratio
print(companies.groupby("Bank")["Debt_Ratio"].mean())

#revenue growth
companies["Revenue_Growth"] = companies.groupby("Bank")["Revenue"].pct_change()
print(companies.groupby("Bank")["Revenue_Growth"].mean())

#want to visualise the profit margin over time
import matplotlib.pyplot as plt
for bank in companies["Bank"].unique():
    subset = companies[companies["Bank"] == bank]
    subset = subset.sort_values("Date")
    plt.plot(subset["Date"], subset["Profit_Margin"], label = bank)

plt.legend()
plt.title("Profit Margin Comparison")
plt.show()
```
